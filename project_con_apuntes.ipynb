#### Exercise 01. Which is the most expensive house in the dataset? (★☆☆)

Print the address and the price of the selected house. For example:

`The house with address General Street Nº5 is the most expensive and its price is 5000000 USD`

```python
import pandas as pd
import numpy as np

highest_price = df['price'].max() #esto busca el valor más alto del level "price" en todo el csv
print(highest_price)

highest_price_row = df.loc[df['price'] == highest_price] #Del highest_price printar toda la fila o row con el highest_price
#highest_price_row es toda la información de la fila con el precio más alto el ds.loc de precio debe ser = a precio mas alto,
#.loc siempre trae la fila, .loc = me imprime el valor o contenido.'''

print(highest_price_row) #me imprime toda la información de la fila con el precio más alto

house_address_high = highest_price_row['address'].values[0] #.values porque la fila me trae una serie de valores que si son correctos,
#el codigo reconoce que la lista tiene toda la info de la fila
#con el [0] le digo que se pare en el primer valor que es el primer valor que cumple con condición del precio
#.values[] me traer el contenido de la dirección en forma de texto

print(house_address_high)

print(f'The house with address {house_address_high} is the most expensive and its price is {highest_price} USD')
```

```python
import pandas as pd
import numpy as np

#MAL!! #print(df.loc["NaN"].index()) Los valores nulos no son strings!!!
 #Los valores nulos NA o NaN no van entre comillas

NAs = pd.isna(df)

df.info() #otra forma de ver donde estan los datos nulos aunque no me piden eso
df.shape #me imprime número total de filas  las columnas que hay en la tabla



# Suponiendo que el dataframe "df" ya está cargado
# Creamos una función para verificar y mostrar los NAs
def print_na_positions(df):
    # Variable para determinar si se encontraron NAs
    found_na = False #lo pongo antes porque la primera condicion de que si no tiene es falso, estoy descartando de entrada todos los que no son nulos!!! 

    # Iterar sobre filas y columnas para encontrar NAs
    for row in range(df.shape[0]): #df.shape le dice que busque en todas las filas y columnas del csv dd índice cero de ambas
        for col in range(df.shape[1]): #df.shape le dice que busque desde la columna con índice 2 (segunda) del csv 
            if pd.isna(df.iloc[row, col]):
                found_na = True
                print(f"True - NA encontrado en fila {row}, columna {col}")
    
    # Si no se encontraron NAs
    if not found_na:
        print("False - No se encontraron NAs en el dataframe")

# Llamar a la función con el dataframe ds
print_na_positions(df)





#Intentos fallidos previos:


#intentos previos

#Ponemos como parámetro en la descripción de la función entre ()
#Quiero que la función detecte nulos en el csv, por eso le paso como parametro mi dataframe

#Solución correcta:




```
